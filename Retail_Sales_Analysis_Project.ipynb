```python
# Use print("messages...") to debug
import csv
import pandas as pd
import matplotlib.pyplot as plt

class RetailSalesAnalyzer:
    def __init__(self):
        self.data = pd.read_csv('/content/retail_sales.csv')
        self.data['Date'] = pd.to_datetime(self.data['Date'])

    def dataCleansing(self):
      self.dropna(inplace=True)

    def total_sales_per_prpduct(self):
      return self.data.groupby('Product')['Sales'].sum()

    def best_selling_product(self):
      return self.total_sales_per_prpduct().sort_values(ascending=False).index[0]

    def avg_daily_sales(self):
      return self.data['Sales'].mean()

    def plot_sales_trend(self):
      self.data.groupby('Date')['Sales'].sum().plot(kind = 'line')
      plt.title('Sales Trend Over Time')
      plt.xlabel('Date')
      plt.ylabel('total Sales')
      return plt.show()

    def plot_sales_per_product(self):
      self.total_sales_per_prpduct().plot(kind='bar')
      plt.title('Sales per Product')
      plt.xlabel('Product')
      plt.ylabel('Total Sales')
      return plt.show()

```

```python
analyzer = RetailSalesAnalyzer()
print('Total sales per Product: \n', analyzer.total_sales_per_prpduct())
print('Best Selling Product: \n', analyzer.best_selling_product())
print('Avg Daily Sales: \n', analyzer.avg_daily_sales())
analyzer.plot_sales_trend()
analyzer.plot_sales_per_product()
```
